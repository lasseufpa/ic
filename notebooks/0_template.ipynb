## Pré-processamento

```python
from sklearn.datasets import load_digits
from sklearn.model_selection import train_test_split

# Carregando o conjunto de dados de dígitos
digits = load_digits()

X = digits.data   # features (imagens achatadas)
y = digits.target # labels (dígitos 0–9)

# Dividir em treino (80%) e teste (20%)
X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.2,
    random_state=2026, # a seed não deve ser modificada
    stratify=y  # mantém a proporção das classes
)

print("Treino:", X_train.shape, y_train.shape)
print("Teste:", X_test.shape, y_test.shape)
```
